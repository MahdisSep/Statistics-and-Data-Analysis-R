```r
set.seed(100)
data <- sample(1:100, 100, replace = TRUE) 

mean_original <- mean(data)
sd_original <- sd(data)
range_original <- range(data)

hist(data, breaks = 10, main = "Histogram of Original Data",
     xlab = "Values", col = "blue", border = "white")

cat("Original Data Statistics:\n")
cat("Mean:", mean_original, "\n")
cat("Standard Deviation:", sd_original, "\n")
cat("Range:", range_original, "\n\n")

```

```r
normalized_range <- (data - min(data)) / (max(data) - min(data))



hist(normalized_range, breaks = 10, main = "Histogram of Range-Normalized Data",
     xlab = "Normalized Values", col = "lightpink", border = "white")

mean_normalized <- mean(normalized_range)
sd_normalized <- sd(normalized_range)
range_normalized <- range(normalized_range)

cat("Range Normalized Data Statistics:\n")
cat("Mean:", mean_normalized, "\n")
cat("Standard Deviation:", sd_normalized, "\n")
cat("Range:", range_normalized, "\n\n")



```

By applying this concept, we observe that all the data have been mapped from their previous range to the interval [0, 1], and the new range of values is now between 0 and 1.


```r
normalized_distribution <- (data - mean(data)) / sd(data)

mean_normalized <- round(mean(normalized_distribution),2)
sd_normalized <- round(sd(normalized_distribution),2)
range_norm <- round(range(normalized_distribution),2)


hist(normalized_distribution, breaks = 10, main = "Histogram of Distribution-Normalized Data",
     xlab = "Normalized Distribution Values", col = "lightblue", border = "white")

cat("Distribution Normalized Data Statistics:\n")
cat("Mean:", mean_normalized, "\n")
cat("Standard Deviation:", sd_normalized, "\n")
cat("Range:", range_norm, "\n\n")
```

By applying this part as well, we observe that the distribution has become a standard normal distribution, with the mean of the data approximately equal to 0 and the standard deviation equal to 1.


Its advantages can be summarized as follows:


1. **Increased accuracy of statistical analyses:**
   In statistical analyses that depend on the scale of variables, normalizing the data prevents variables with larger ranges from dominating the results.


2. **Feature comparability:**
   When variables have different scales (for example, weight in kilograms, height in centimeters, and age as a number between 1 and 120), normalization allows these values to become comparable.


3. **Facilitated convergence:**
   In optimization algorithms such as gradient descent, normalized data help the algorithm converge faster, because values within similar ranges lead to more stable gradients.


4. **Improved model performance:**
   Many machine learning algorithms are sensitive to the scale of features. Normalizing the data prevents the model from becoming overly sensitive to features with larger ranges.


Its disadvantages can also be summarized as follows:


1. **Loss of relative information:**
   In some cases, normalization can diminish the relative information between values, especially when the data have specific physical meanings.


2. **Sensitivity to noise:**
   If the data contain a lot of noise, normalization can amplify it, since all values are compressed into a smaller scale.


Overall, range normalization reduces the interpretability of data for the reasons mentioned in points 1 and 2. This is because when a data point originally represented noise, it used to have a significant distance from other data points â€” but by mapping all values to the [0, 1] range, this meaningful distance is lost.
