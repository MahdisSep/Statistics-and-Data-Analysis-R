```r
library(ggplot2)
library(dplyr)

roll_dice <- function(num_rolls) {
  rolls <- sample(1:6, num_rolls, replace = TRUE)
  freq_rolls <- table(rolls)
  freq_percentage <- (freq_rolls / num_rolls) * 100
  data.frame(dice_num = as.numeric(names(freq_rolls)),
             Frequency = as.numeric(freq_percentage),
             Rolls = as.factor(num_rolls))
}

data_100 <- roll_dice(100)
data_1000 <- roll_dice(1000)
data_10000 <- roll_dice(10000)

all_data <- bind_rows(data_100, data_1000, data_10000)
values = c("skyblue", "pink", "lightgreen")

ggplot(all_data, aes(x = dice_num, y = Frequency, fill = Rolls)) +
  geom_bar(stat = "identity", position = "dodge") +
  scale_fill_manual(values = c("skyblue", "pink", "lightgreen")) +
  labs(title = "Frequency Percentage of Dice Rolls",
       x = "Dice number", y = "Frequency Percentage (%)",
       fill = "Number of Rolls") +
  theme_minimal()
```

```r
library(ggplot2)
library(dplyr)

roll_dice <- function(num_rolls) {
  set.seed(1)
  rolls <- sample(1:6, num_rolls, replace = TRUE)
  freq_rolls <- table(rolls)
  freq_percentage <- (freq_rolls / num_rolls) * 100
  data.frame(dice_num = as.numeric(names(freq_rolls)),
             Frequency = as.numeric(freq_percentage),
             Rolls = as.factor(num_rolls))
}

data_100 <- roll_dice(100)
data_1000 <- roll_dice(1000)
data_10000 <- roll_dice(10000)

all_data <- bind_rows(data_100, data_1000, data_10000)

ggplot(all_data, aes(x = dice_num, y = Frequency, fill = Rolls)) +
  geom_bar(stat = "identity", position = "dodge") +
  scale_fill_manual(values = c("skyblue", "pink", "lightgreen")) +
  labs(title = "Frequency Percentage of Dice Rolls",
       x = "Dice number", y = "Frequency Percentage (%)",
       fill = "Number of Rolls") +
  theme_minimal()
```

**C**

In this section, we observe that setting the seed causes the same random numbers to be generated every time we run the cell, meaning no new random numbers are produced. However, if we do not set the seed function, we can generate new random numbers each time we run the code.

From these two parts, we can also conclude that when the number of experiment repetitions is small, the frequency percentage of the dice outcomes is not equal â€” one number may appear more frequently than another. But as we increase the number of repetitions, we notice that the frequency percentages of different dice outcomes become closer to each other (it is expected that the frequency of each number approaches 1/6).

In fact, as the number of trials increases, the data tend to become more normalized and evenly distributed. This experiment emphasizes the importance of sample size and repeatability in statistical experiments. Overall, as the number of dice throws increases, the distribution of results approaches the ideal uniform distribution.
